```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_wine
from mpl_toolkits.mplot3d import Axes3D

# Load dataset
wine = load_wine(as_frame=True)
df = wine.frame

# ------------------------------
# Q1. Line graph (alcohol vs malic_acid)
plt.plot(df['alcohol'], df['malic_acid'])
plt.xlabel("Alcohol")
plt.ylabel("Malic Acid")
plt.title("Alcohol vs Malic Acid (Line Plot)")
plt.show()

# ------------------------------
# Q2. Boxplot of flavanoids grouped by wine class
sns.boxplot(x="target", y="flavanoids", data=df)
plt.title("Flavanoids by Wine Class")
plt.show()

# ------------------------------
# Q3. 2D scatterplot (alcohol vs color_intensity)
plt.scatter(df['alcohol'], df['color_intensity'], c=df['target'])
plt.xlabel("Alcohol")
plt.ylabel("Color Intensity")
plt.title("Alcohol vs Color Intensity (Colored by Class)")
plt.show()

# ------------------------------
# Q4. 3D scatterplot (alcohol, color_intensity, hue=wine class)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(df['alcohol'], df['color_intensity'], df['hue'], c=df['target'])
ax.set_xlabel("Alcohol")
ax.set_ylabel("Color Intensity")
ax.set_zlabel("Hue")
plt.title("3D Scatterplot: Alcohol, Color Intensity, Hue")
plt.show()

# ------------------------------
# Q5. Pairplot with 4 numerical columns
sns.pairplot(df[['alcohol','malic_acid','flavanoids','color_intensity','target']], hue='target')
plt.show()

# ------------------------------
# Q6. Limitations of pairplots
"""
1. Pairplots become cluttered for high-dimensional datasets (too many plots).
   -> Alternative: PCA or t-SNE for dimensionality reduction.
2. They only show pairwise relationships, ignoring complex multivariate interactions.
   -> Alternative: Correlation heatmap or multivariate plots (3D plots, parallel coordinates).
"""

# ------------------------------
# Q7. Histogram of alcohol + PDF
sns.histplot(df['alcohol'], kde=True, stat="density")
plt.xlabel("Alcohol")
plt.title("Alcohol Distribution with PDF")
plt.show()

# ------------------------------
# Q8. Univariate analysis of malic_acid
sns.kdeplot(df['malic_acid'], fill=True)
plt.xlabel("Malic Acid")
plt.title("PDF of Malic Acid")
plt.show()

# Comment: Distribution is right-skewed with most values concentrated near lower range.

# ------------------------------
# Q9. CDF of proline + % < 800
sorted_proline = np.sort(df['proline'])
cdf = np.arange(1, len(sorted_proline)+1) / len(sorted_proline)

plt.plot(sorted_proline, cdf)
plt.xlabel("Proline")
plt.ylabel("CDF")
plt.title("CDF of Proline")
plt.show()

percent_less_800 = (df['proline'] < 800).mean() * 100
print(f"Percentage of samples with proline < 800: {percent_less_800:.2f}%")
```
